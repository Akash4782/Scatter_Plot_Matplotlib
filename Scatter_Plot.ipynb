Scatter Plot

```python
import matplotlib.pyplot as plt 
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number", fontsize =15)
plt.scatter(x,y, color = "r")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
colors = ["r","y","g","r","y","g","r"]
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number", fontsize =15)
plt.scatter(x,y, c = colors)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
colors = ["r","y","g","r","y","g","r"]
sizes = [100,200,400,205,202,150,71]
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number", fontsize =15)
plt.scatter(x,y, c = colors , s = sizes)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
colors = ["r","y","g","r","y","g","r"]
sizes = [100,200,400,205,202,150,71]
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number", fontsize =15)
plt.scatter(x,y, c = colors , s = sizes , alpha = 0.5)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
colors = ["r","y","g","r","y","g","r"]
sizes = [250,500,450,605,700,350,600]
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number", fontsize =15)
plt.scatter(x,y, c = colors , s = sizes , marker ="*" , edgecolor = "black" , linewidth = 2)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
colors = [10,29,30,27,45,67,34]
sizes = [100,200,400,205,202,150,71]
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number", fontsize =15)
plt.scatter(x , y , c=colors , cmap = "viridis" , s = sizes)
t = plt.colorbar()
t.set_label("ColorBar",fontsize = 10)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
y1 = [3,4,2,7,6,4,9]
colors = [10,29,30,27,45,67,34]
sizes = [100,200,400,205,202,150,71]
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number", fontsize =15)
plt.scatter(x , y , c=colors , cmap = "viridis" , s = sizes)
plt.scatter(x , y1 , color = "r" ,  s = sizes)
t = plt.colorbar()
t.set_label("ColorBar",fontsize = 10)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
y1 = [3,4,2,7,6,4,9]
colors = [10,29,30,27,45,67,34]
sizes = [100,200,400,205,202,150,71]
plt.title("Scatter Plot", fontsize = 20)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("Number", fontsize =15)
plt.scatter(x , y , c=colors , cmap = "viridis" , s = sizes ,marker = "*", edgecolor = "black" , linewidth = 1)
plt.scatter(x , y1 , color = "r" ,  s = sizes, marker = "*", edgecolor = "black" , linewidth = 1)
t = plt.colorbar()
t.set_label("ColorBar",fontsize = 10)
plt.show()
```
